```python
import os
import numpy as np
import scipy as sp
import librosa
import matplotlib.pyplot as plt
import pywt
import sounddevice as sd
import sympy
```

```python
#1.5.1 
c=340 #speed of light
thita_s=90 #dedomeno
f=2000 #dedomeno
w=2*np.pi*f
d=0.04 #dedomeno
N_array=[4,8,16]
thita=np.linspace(0,180,100) #gwnia se rad
color_array=['-r','-b','-g'] #tria diaforetika xrwmata
legend_array=['N=4','N=8','N=16'] #red->N=4 // blue->N=8 // green-> N=16 
for i in range (0,3):
    N=N_array[i] #epilogi N
    color=color_array[i] #epilogi xrwmatos grafikis
    B_upper=(1)*np.sin(((N*w)/(2*c))*d*(np.cos(np.pi*thita/180)- np.cos(np.pi*thita_s/180))) #arithmitis
    B_lower=(N)*np.sin(((1*w)/(2*c))*d*(np.cos(np.pi*thita/180)- np.cos(np.pi*thita_s/180)))+0.0000001 #paronomastis
    B=B_upper/B_lower #delay-and-sum beam pattern (sxesi 19)
    plt.plot(thita,np.log10(abs(B)+0.00000001),color)
    plt.xlabel('Γωνία(Μοίρες)')
    plt.ylabel('Beam-Pattern Amplitude(dB)')
    #plt.title(legend_array[i]) #oi grammes se sxolia einai gia na ginoun xwrista oi grafikes (anti gia oles mazi)
    plt.legend(legend_array)
    #plt.show()
plt.show()    


#sumperasma-> auksisi N odigei se pio amesi apokopi 
```

```python
#1.5.2
c=340 #speed of light
thita_s=90 #dedomeno
f=2000 #dedomeno
w=2*np.pi*f
N=8 #dedomeno
d_array=[0.04,0.08,0.16]
thita=np.linspace(0,180,100) #gwnia se rad
color_array=['-r','-b','-g'] #tria diaforetika xrwmata
legend_array=['d=4cm','d=8cm','d=16cm'] #red->N=4 // blue->N=8 // green-> N=16 
for i in range (0,3):
    d=d_array[i] #epilogi d
    color=color_array[i] #epilogi xrwmatos grafikis
    B_upper=(1)*np.sin(((N*w)/(2*c))*d*(np.cos((np.pi*thita)/180)- np.cos((np.pi*thita_s)/180))) #arithmitis
    B_lower=(N)*np.sin(((1*w)/(2*c))*d*(np.cos((np.pi*thita)/180)- np.cos((np.pi*thita_s)/180)))+0.000000001 #paronomastis
    B=B_upper/B_lower #delay-and-sum beam pattern (sxesi 19)
    plt.plot(thita,np.log10(abs(B)+0.00000001),color)
    plt.xlabel('Γωνία(Μοίρες)')
    plt.ylabel('Beam-Pattern Amplitude(dB)')
    #plt.title(legend_array[i]) #oi grammes se sxolia einai gia na ginoun xwrista oi grafikes (anti gia oles mazi)
    plt.legend(legend_array)
    #plt.show()
plt.show()    

#sumperasma: auksisi tou d odigei se pio apotomo kopsimo 
```

```python
#1.5.3
c=340 #speed of light
thita_s=90 #dedomeno
N=8 #dedomeno
d=0.08 #dedomeno
f=np.linspace(0,8000,100) 
w=2*np.pi*f#gwnia se rad
thita_array=[0,40,60]
color_array=['-r','-b','-g'] #tria diaforetika xrwmata
legend_array=['thita=0 deg','thita=40 deg','thita=60 deg']  
for i in range (0,3):
    thita=thita_array[i] #epilogi N
    color=color_array[i] #epilogi xrwmatos grafikis
    B_upper=(1)*np.sin(((N*w)/(2*c))*d*(np.cos(np.pi*thita/180)- np.cos(np.pi*thita_s/180))) #arithmitis
    B_lower=(N)*np.sin(((1*w)/(2*c))*d*(np.cos(np.pi*thita/180)- np.cos(np.pi*thita_s/180)))+0.000000001 #paronomastis
    B=B_upper/B_lower #delay-and-sum beam pattern (sxesi 19)
    plt.plot(f,np.log10(abs(B)+0.00000001),color)
    plt.xlabel('Frequency(Hz)')
    plt.ylabel('Beam-Pattern Amplitude(dB)')
    plt.title(legend_array[i]) #oi grammes se sxolia einai gia na ginoun xwrista oi grafikes (anti gia oles mazi)
    #plt.legend(legend_array)
    plt.show()
#plt.show()    

#sumperasma: auksisi thita -> auksisi suxnotitas megistopoihshs
#thorivos diaforetikis kateuthinsiis kovetai(apotelesma twn 1.5.1+1.5.2)
```
