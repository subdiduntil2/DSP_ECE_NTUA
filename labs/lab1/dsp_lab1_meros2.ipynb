```python
import os
import numpy as np
import scipy as sp
import librosa
import matplotlib.pyplot as plt
import pywt
import sounddevice as sd
import sympy
```

```python
#2.1.a
Fs=1000
Ts=0.001
N=2000     #Fs=1000Hz => Ts=0.001sec => N=Tol/Ts=(2-0)sec/0.001sec => N=2000 samples
t1=np.linspace(0,2,N)
x1 = 2*np.cos(2*np.pi*70*t1) + 3*np.sin(2*np.pi*100*t1)+ 0.1*np.random.normal(size=(2000,)) #size=2000=N
plt.plot(t1,x1)
plt.xlabel('Time(sec)')
plt.ylabel('Amplitude(V)')
plt.title('Sampled Input Signal A')
plt.show()
```

```python
#2.1b
%matplotlib inline
s1 = librosa.stft(x1,n_fft=40,hop_length=20,win_length=40)#win_length=0.04sec*1000Hz=40=n_fft #hop_length=win_length/2
print(s1.shape) #(1025,2) #number of horizontal/vertical bins is determined by n_fft/hop_length
t1 = np.linspace(0,2,101) #(0,Tol,y_axis_of_s1) 
f1 = np.linspace(0,1000/2,21) #(0,Fs/2,x_axis_of_s1)
#plt.pcolormesh(t1,f1,np.abs(s1))
plt.pcolormesh(t1,f1,np.log10(np.abs(s1)+0.00001)) #log scale for better output
plt.xlabel('Time (sec)')
plt.ylabel('Frequency(Hz)')
plt.title('STFT of the input signal A with win_length=0.04sec')
plt.show()
```

```python
#2.1c for 0.08
%matplotlib inline
s1 = librosa.stft(x1,n_fft=80,hop_length=40,win_length=80)#win_length=0.08sec*1000Hz=80=n_fft #hop_length=win_length/2
print(s1.shape) #(1025,2)#number of horizontal/vertical bins is determined by n_fft/hop_length
t1 = np.linspace(0,2,51) #(0,Tol,y_axis_of_s1) 
f1 = np.linspace(0,1000/2,41) #(0,Fs/2,x_axis_of_s1)
#plt.pcolormesh(t1,f1,np.abs(s1))
plt.pcolormesh(t1,f1,np.log10(np.abs(s1)+0.00001)) #log scale for better output
plt.xlabel('Time (sec)')
plt.ylabel('Frequency(Hz)')
plt.title('STFT of the input signal A with win_length=0.08sec')
```

```python
#2.1c for 0.16
%matplotlib inline
s1 = librosa.stft(x1,n_fft=160,hop_length=80,win_length=160)#win_length=0.16sec*1000Hz=160=n_fft#hop_length=win_length/2
print(s1.shape) #(1025,2) #number of horizontal/vertical bins is determined by n_fft/hop_length
t1 = np.linspace(0,2,26) #(0,Tol,y_axis_of_s1) 
f1 = np.linspace(0,1000/2,81) #(0,Fs/2,x_axis_of_s1)
#plt.pcolormesh(t1,f1,np.abs(s1))
plt.pcolormesh(t1,f1,np.log10(np.abs(s1)+0.00001)) #log scale for better output
plt.xlabel('Time (sec)')
plt.ylabel('Frequency(Hz)')
plt.title('STFT of the input signal A with win_length=0.16sec')
#sumperasmata 2.1b+2.1c: auksisi win_length -> kaluteros prosdiorismos fasmatos tou simatos gurw apo tin zwni twn 200Hz
#kai xeiroteros xronikos prosdiorismos
#gia win=0.16sec exoume kai diaforopoiisi twn 2 kuriarxwn suxnotitwn se 2 ypolwride
```

```python
#2.1d for CWT(t,f)
sa = np.power(2,np.linspace(1,6,16*5)) #scale [1,64]-> freqs [500,15.65] #16*5 samples -> 16 samples/octave
print(f1)
coefs,freqs = pywt.cwt(x1,sa,'cmor3.0-1.0')
t1 = np.linspace(0,2,N) #(Tmin,Tmax,N)
f1 = freqs*Fs #get the actual freq in Hz (not the scaled one)
#plt.pcolormesh(t1,f1,np.abs(coefs))
plt.pcolormesh(t1,f1,np.log10(np.abs(coefs)+0.00001)) #log scale for better output
plt.xlabel('Time (sec)')
plt.ylabel('Frequency(Hz)')
plt.title('CWT(t,f) of the input signal A')

```

```python
#2.1d for CWT(t,s)
#plt.pcolormesh(t1,sa,np.abs(coefs))
plt.pcolormesh(t1,sa,np.log10(abs(coefs)+0.00001)) #log scale for better output
plt.xlabel('Time (sec)')
plt.ylabel('Scale')
plt.title('CWT(t,s) of the input signal A')
#sumperasmata 2.1e: paratiroume oti o CWT kanei kali suxnotiki analusi mono stin perioxi endiaferontos (gurw sta 100Hz)
#mias kai xrisimopoiei metavalomena se mikos parathira, fainetai apo tin pio smooth kitrini lwrida sta 100Hz
#adunamia diaxwrismou twn 2 suxnotitwn (mazi se mia lwrida) se antithesi me STFT
```

```python
#2.2a
Fs=1000
Ts=0.001
N=2000     #Fs=1000Hz => Ts=0.001sec => N=Tol/Ts=(2-0)sec/0.001sec => N=2000 samples
t2=np.linspace(0,2,N) 
dirac= 1.7*(sp.signal.unit_impulse(N,750)-sp.signal.unit_impulse(N,900)) #palmoi sta 0.75(simeio 750) kai 0.9(simeio 900)
x2 = 1.5*np.cos(2*np.pi*80*t2) + 0.15*np.random.normal(size=(2000,)) + dirac
plt.plot(t2,x2)
plt.xlabel('Time(sec)')
plt.ylabel('Amplitude(V)')
plt.title('Sampled Input Signal B')
```
